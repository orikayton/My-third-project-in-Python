C11-C16 checks the relationship between the coffee attribute scores (correlation).
(negative relationship/weak relationship/medium relationship/strong relationship).
1) It is not possible to see a dependency between the 'uniformity' attribute in relation to the other attributes{ [see C11 and C}2]. (also in C13). In addition, this can also be seen according to C16. An explanation for the uniformity property values being the same in all tests (for an explanation, see C17)
2) You can see a single example - a correlation test between the attribute 'Aroma' and 'Aftertaste', see C14
3) Strong correlations:
- between 'Aftertaste' and 'Flavor' and 'Balance' (0.9)
- Between 'Acidity' and 'Aftertaste' and 'Flavor' (0.8)
- between 'Aroma' and 'Aftertaste' and 'Flavor' (0.8)
- Medium relationship between 'Aroma' and 'Body' (0.6) and between 'Acidity' (0.7)
Summary: It can be said that most of the coffee properties that are measured/tested are dependent on each other and it is possible that one property that is tested as "low" is enough to affect the quality of the coffee
